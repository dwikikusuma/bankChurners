### lihat uniqe value pada dataset

```python
def desc(x):
    temp = []
    d =[]
    for i in x:
        a = data[i].unique()
        c = data[i].nunique()
        b = str(a)
        temp.append(b)
        d.append(c)
    res = pd.DataFrame(temp,index=x,columns=['unique value'])
    res['count'] = d
    return res
```

```python
cek = 'Attrition_Flag','Gender','Dependent_count', 'Education_Level', 'Marital_Status','Income_Category','Card_Category'
desc(cek)
```

### Churn By gendr

```python
attr_g = data.groupby('Attrition_Flag')['Gender'].value_counts().to_frame().rename(columns={'Gender': 'Freq'}).reset_index().sort_values('Attrition_Flag')
plt.figure(figsize=(10,10))
sns.barplot(data=attr_g,x="Attrition_Flag",y='Freq',hue='Gender')
plt.title("Attrition By Gender")
plt.legend(loc=0)
plt.show()
```

Gender pengguna yang melakukan churn tidak jauh berbeda

```python
attr_g = attr_g.pivot_table('Freq',['Attrition_Flag','Gender'])
attr_g
```
